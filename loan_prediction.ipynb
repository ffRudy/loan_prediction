```python
import pandas as pd

#读取贷款数据集
df = pd.read_csv("/content/Loan_Default.csv")

#去除Upfront_charges,Interest_rate_spread,rate_of_interest三列
df = df.drop(['Upfront_charges', 'Interest_rate_spread', 'rate_of_interest'], axis=1)
df = df.drop(['year'], axis=1)
# 统计每列的缺失值比例
col_missing = df.isnull().sum() / len(df)
col_missing = col_missing.sort_values(ascending=False)  # 按缺失值数量排序

# 绘制柱状图
fig, ax = plt.subplots(figsize=(10, 6))
# plt.bar(x=col_missing.index, height=col_missing)
# plt.xticks(rotation=90)
# plt.xlabel("Columns")
# plt.ylabel("Missing Value Ratio")
# plt.title("Missing Value Ratio in Loan Data")
ax.barh(y=col_missing.index, width=col_missing)
ax.set_ylabel("Columns")
ax.set_xlabel("Missing Value Ratio")
ax.set_title("Missing Value Ratio in Loan Data")
plt.show()
```

```python
# 对于数据类型的列使用平均值填充缺失值
mean_values = df.select_dtypes(include=['float64', 'int64']).mean()
df = df.fillna(mean_values)

# 对于 object 类型的列使用出现次数最多的值进行填充
mode_values = df.select_dtypes(include=['object']).mode().iloc[0]
df = df.fillna(mode_values)
```

```python
col_missing = df.isnull().sum() / len(df)
col_missing = col_missing.sort_values(ascending=False)  # 按缺失值数量排序

# 绘制柱状图
fig, ax = plt.subplots(figsize=(10, 6))
# plt.bar(x=col_missing.index, height=col_missing)
# plt.xticks(rotation=90)
# plt.xlabel("Columns")
# plt.ylabel("Missing Value Ratio")
# plt.title("Missing Value Ratio in Loan Data")
ax.barh(y=col_missing.index, width=col_missing)
ax.set_ylabel("Columns")
ax.set_xlabel("Missing Value Ratio")
ax.set_title("Missing Value Ratio in Loan Data")
plt.show()
```

```python
import seaborn as sns

# 选择需要进行相关性分析的列并构造一个新的 DataFrame
df_corr = pd.concat([df.select_dtypes(include=['float64', 'int64']),
                     pd.get_dummies(df['Status'])],
                    axis=1)

# 计算相关性矩阵
corr = df_corr.corr().round(1)

# 绘制热力图
sns.heatmap(corr, vmin=-1, vmax=1, cmap='coolwarm', annot=True, square=True)
```
